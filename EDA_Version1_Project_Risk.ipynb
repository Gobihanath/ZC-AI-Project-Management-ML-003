```python
irrelevant = [
    'Complexity_Score','Change_Request_Frequency','Team_Turnover_Rate',
    'Communication_Frequency','Geographical_Distribution','Market_Volatility',
    'Cross_Functional_Dependencies','Technical_Debt_Level','Seasonal_Risk_Factor'
]
```

```python
df.head()
```

```python
present = [c for c in irrelevant if c in df.columns]
print("Columns found and will be dropped:", present)
```

```python
df = df.drop(columns=present, errors='ignore')
print("New shape:", df.shape)
```

```python
for col in df.columns:
    if 'date' in col.lower():
        try:
            df[col] = pd.to_datetime(df[col])
            print("Converted to datetime:", col)
        except Exception:
            pass
```

```python

for col in df.columns:
    if df[col].dtype == 'object':

        try:
            df[col] = pd.to_numeric(df[col].str.replace(',',''), errors='ignore')
        except Exception:
            pass
display(df.dtypes)
```

```python
# Compute % missing and decide strategy
missing_pct = df.isnull().mean().sort_values(ascending=False) * 100
display(missing_pct[missing_pct > 0].round(2))

cols_to_drop = missing_pct[missing_pct > 50].index.tolist()
print("Columns with >50% missing (consider dropping):", cols_to_drop)
```

```python
# Fill numeric missing with median
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
for c in num_cols:
    if df[c].isnull().any():
        df[c].fillna(df[c].median(), inplace=True)
```

```python
# Fill categorical with mode
cat_cols = df.select_dtypes(include=['object','category']).columns.tolist()
for c in cat_cols:
    if df[c].isnull().any():
        df[c].fillna(df[c].mode().iloc[0], inplace=True)

print("Missing values after simple imputation:")
display(df.isnull().sum().sort_values(ascending=False).head(20))
```

```python
# 7. Prepare numeric and categorical lists
numeric_df = df.select_dtypes(include=[np.number]).copy()
categorical_cols = df.select_dtypes(include=['object','category']).columns.tolist()
print("Numeric columns:", numeric_df.columns.tolist())
print("Categorical columns:", categorical_cols)
```

```python
# Correlation matrix
corr = numeric_df.corr(method='pearson')
plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm', cbar=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Extract top correlated pairs (absolute value, exclude self-pairs)
corr_pairs = corr.abs().unstack().reset_index()
corr_pairs.columns = ['feature_1','feature_2','abs_corr']
```

```python
# remove self correlations
corr_pairs = corr_pairs[corr_pairs['feature_1'] != corr_pairs['feature_2']]

corr_pairs['pair'] = corr_pairs.apply(lambda r: tuple(sorted([r.feature_1, r.feature_2])), axis=1)
corr_pairs = corr_pairs.drop_duplicates(subset=['pair']).sort_values('abs_corr', ascending=False)
display(corr_pairs[['feature_1','feature_2','abs_corr']].head(15))
```

```python
# Visualize top correlated pairs
top_pairs = corr_pairs[['feature_1','feature_2']].head(15).values.tolist()
for f1, f2 in top_pairs:
    plt.figure(figsize=(6,4))
    sns.scatterplot(data=df, x=f1, y=f2)
    plt.title(f"Scatter: {f1} vs {f2} (corr={corr.loc[f1,f2]:.2f})")
    plt.show()
```
